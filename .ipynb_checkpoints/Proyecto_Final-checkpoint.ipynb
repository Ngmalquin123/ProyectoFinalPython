# Introduccion al dataset

El dataset proviene del Servicio de Rentas Internas del Ecuador (SRI) y contiene datos de ventas y compras declaradas en el Formulario 104 (IVA) desde 2019 hasta 2024.

Ecuador esta organizado en 24 provincias agrupadas en 4 regiones naturales: Costa, Sierra, Amazonia e Insular. Cada registro contiene informacion del contribuyente a nivel de canton (division politico-administrativa menor a la provincia).

# Objetivo

Analizar los datos de ventas y compras en Ecuador desde 2019 hasta 2024, utilizando información del Servicio de Rentas Internas (SRI) - Formulario 104.

# Objetivos Específicos

1- Entender que sectores economicos tienen mayor dinamismo y crecimiento.

2- Identificar provincias con alto potencial comercial no explotado.

3- Detectar la estacionalidad del comercio para optimizar decisiones de negocio.

4- Comparar la evolucion pre y post COVID-19 en el comercio ecuatoriano.

5- Encontrar sectores con bajo volumen pero alto crecimiento (nichos de mercado).

# Enfoque de este trabajo

Se busca que los datos esten completamente limpios y enriquecidos con variables adicionales que permitan el analisis. El dataset tiene buena completitud desde el inicio pero requiere normalizacion.

Para esto se emplearan varios pasos:

1- Normalizacion y limpieza de los campos (texto, numericos, duplicados)

2- Enriquecimiento con la region geografica de cada provincia

3- Enriquecimiento con el nombre del sector economico CIIU

4- Creacion de variables derivadas: TOTAL_VENTAS calculado, margen, trimestre

5- Union de los 6 archivos anuales en un unico dataset consolidado

Una vez procesada la data se aplicaran analisis matematicos para calcular tendencias, crecimiento por sector y provincia, estacionalidad y deteccion de oportunidades.


# Primera impresion del dataset

Los dataset contienen aprox 297k entradas entre 2019 y 2024. Los campos AÑO, MES, PROVINCIA y CANTON estan siempre presentes.

Las columnas de ventas y compras estan bien estructuradas segun el diccionario oficial del SRI.

El mayor desafio sera normalizar los textos de provincias y sectores para poder agrupar correctamente.

###  Diccionario de Datos Detallado:

**Variables que indican Temporalidad:**
- `AÑO`: Año fiscal de la declaración (2019-2024)
- `MES`: Mes fiscal (1-12). Indica el período de la transacción

**Variables Geográficas:**
- `PROVINCIA`: División administrativa de primer nivel (24 provincias)
- `CANTON`: División administrativa de segundo nivel (~221 cantones)
- `REGION` (derivada): Región natural (Costa, Sierra, Amazonía, Insular)

**Variables de Clasificación Económica:**
- `CODIGO_SECTOR_N1`: Código CIIU 4.0 Nivel 1 (actividad económica)
- `SECTOR_NOMBRE` (derivada): Nombre legible del sector

**Variables Monetarias (USD):**
- `VENTAS_NETAS_TARIFA_12`: Ventas con IVA 12% (bienes y servicios gravados)
- `VENTAS_NETAS_TARIFA_0`: Ventas con IVA 0% (productos básicos, exportaciones)
- `EXPORTACIONES`: Ventas al exterior (exentas de IVA)
- `COMPRAS_NETAS_TARIFA_12`: Compras con IVA 12%
- `COMPRAS_NETAS_TARIFA_0`: Compras con IVA 0%
- `IMPORTACIONES`: Compras del exterior
- `COMPRAS_RISE`: Compras a contribuyentes del Régimen Simplificado
- `TOTAL_COMPRAS`: Suma de todas las compras
- `TOTAL_VENTAS`: Suma de todas las ventas

**Variables Derivadas:**
- `MARGEN`: Diferencia entre TOTAL_VENTAS y TOTAL_COMPRAS (rentabilidad bruta)
- `TRIMESTRE`: Agrupación trimestral (Q1, Q2, Q3, Q4)
- `MES_NOMBRE`: Nombre del mes en español

###  Notas Importantes:
- Los valores están en **dólares estadounidenses** (USD)
- Ecuador usa el dólar como moneda oficial desde el año 2000
- Los datos representan **declaraciones agregadas** por sector y ubicación, no transacciones individuales
- Formato numérico original: coma como separador decimal (ej: 1.234,56)

# FASE 1: Descarga y Carga de Datos

Descargaremos los datasets de ventas y compras del SRI para los años 2019-2024.

####  IMPORTANTE: Descarga Manual de Datos

Debido a que el portal de datos abiertos del gobierno requiere navegación web interactiva, se debe descargar los archivos manualmente:

**Pasos para descargar:**

1. **Visita el portal del SRI**: https://www.sri.gob.ec/datasets

2. **Buscar la sección "Ventas y Compras"**

3. **Descarga de los archivos CSV** para los años:
   - SRI_Ventas_2024.csv 
   - SRI_Ventas_2023.csv
   - SRI_Ventas_2022.csv
   - SRI_Ventas_2021.csv
   - SRI_Ventas_2020.csv
   - SRI_Ventas_2019.csv

4. **Guardar los archivos** en la carpeta `data/`


```python
# Crear directorio para guardar datasets si no existe
data_dir = "data"
if not os.path.exists(data_dir):
    os.makedirs(data_dir)
    print(f"Directorio '{data_dir}' creado")
else:
    print(f"Directorio '{data_dir}' ya existe")
```

```python
## Verificar archivos descargados si ya existen o falta alguno y cuantos registros tiene

anios = [2019, 2020, 2021, 2022, 2023, 2024]

dataframes = []

for anio in anios:
    filepath = osp.join(data_dir, f'sri_ventas_{anio}.csv')
    for enc in ['utf-8', 'latin-1', 'cp1252']:
        try:
            df = pd.read_csv(filepath,sep='|', encoding=enc, low_memory=False)
            dataframes.append(df)
            print(f'Cargado {anio}: {len(df):,} registros')
            break
        except UnicodeDecodeError:
            continue

# Unir todos los años
sri_dataset = pd.concat(dataframes, ignore_index=True)
print(f'\nDataset completo: {len(sri_dataset):,} registros')
```


# FASE 2: Limpieza y Normalización de Datos

En esta fase:
1. Unificaremos los datasets de todos los años
2. Normalizaremos nombres de columnas
3. Limpiaremos valores faltantes
4. Estandarizaremos formatos

```python
# Normalizar nombres de columnas: quitar acentos, espacios y convertir a mayusculas
def normalizar_col(texto):
    texto = str(texto).replace('ñ', 'ni').replace('Ñ', 'NI')
    texto = normalize('NFKD', texto).encode('ASCII', 'ignore').decode()
    return texto.strip().upper().replace(' ', '_')

sri_dataset.columns = [normalizar_col(c) for c in sri_dataset.columns]
print('Columnas normalizadas:')
print(sri_dataset.columns.tolist())
```

```python
# Eliminar duplicados completos
antes = len(sri_dataset)
sri_dataset = sri_dataset.drop_duplicates()
print(f'Duplicados eliminados: {antes - len(sri_dataset):,}')
print(f'Registros restantes:   {len(sri_dataset):,}')
```

```python
# 1. Definimos qué valores "ensucian" la completitud
valores_nulos = ['ND', 'NAN', 'N/A', '', 'NONE', 'N_D']

# 2. Creamos una copia temporal para el cálculo o convertimos el dataset
# (Usamos .replace para que Pandas reconozca estos textos como nulos)
sri_dataset_temp = sri_dataset.replace(valores_nulos, np.nan)


completitud = pd.DataFrame({
    'Datos': sri_dataset_temp.notna().sum(),
    'Datos_Nulos': sri_dataset_temp.isna().sum(),
    'Completitud_%': (sri_dataset_temp.notna().sum() / len(sri_dataset_temp) * 100).round(2)
})


display(completitud.sort_values('Completitud_%'))
```

```python
# Normalizar texto en columnas categoricas: quitar acentos, espacios y mayusculas
def limpiar_texto(texto):
    # 1. Si ya es nulo de Pandas, retornar NaN
    if pd.isna(texto):
        return np.nan
    
    # 2. Normalizar: quitar acentos y pasar a mayúsculas
    texto_limpio = normalize('NFKD', str(texto)).encode('ASCII', 'ignore').decode()
    texto_limpio = texto_limpio.strip().upper()
    
    
    
    if texto_limpio in valores_nulos:
        return np.nan  # O podrías retornar "NO ESPECIFICADO"
        
    return texto_limpio

# Aplicar la función
for col in ['PROVINCIA', 'CANTON', 'CODIGO_SECTOR_N1']:
    if col in sri_dataset.columns:
        sri_dataset[col] = sri_dataset[col].apply(limpiar_texto)

print('Provincias únicas encontradas (limpias):', sri_dataset['PROVINCIA'].nunique())
print(sorted(sri_dataset['PROVINCIA'].dropna().unique()))
```

```python
# Convertir columnas numericas y rellenar NaN con 0
cols_numericas = [
    'VENTAS_NETAS_TARIFA_12', 'VENTAS_NETAS_TARIFA_0', 'EXPORTACIONES',
    'COMPRAS_NETAS_TARIFA_12', 'COMPRAS_NETAS_TARIFA_0', 'IMPORTACIONES',
    'COMPRAS_RISE', 'TOTAL_COMPRAS', 'TOTAL_VENTAS'
]

#  Normalizamos tu lista de columnas para que coincida con el dataset ya normalizado
cols_numericas = [normalizar_col(c) for c in cols_numericas]
for col in cols_numericas:
    if col in sri_dataset.columns:
        sri_dataset[col] = (sri_dataset[col]
            .astype(str)
            .str.replace(',', '.', regex=False)
            .str.replace(r'[^0-9.-]', '', regex=True) 
        )
        sri_dataset[col] = pd.to_numeric(sri_dataset[col], errors='coerce')


completeness_ventas = (sri_dataset['TOTAL_VENTAS'].notna().mean() * 100)
print(f'Completitud REAL de TOTAL_VENTAS (antes de imputar 0): {completeness_ventas:.2f}%')

# 3. Ahora sí, rellenamos con 0 para poder operar matemáticamente
sri_dataset[cols_numericas] = sri_dataset[cols_numericas].fillna(0)

# 4. Verificamos el resultado final
display(sri_dataset[cols_numericas].describe())
        
```

```python

# VALIDACIONES DE CALIDAD (CRISP-ML(Q))


print('  REPORTE DE ASEGURAMIENTO DE CALIDAD DE DATOS')


# 1. Validar rango de años
anios_validos = sri_dataset['ANIO'].between(2019, 2024)
anios_invalidos = (~anios_validos).sum()
print(f'\n1️VALIDACIÓN DE AÑOS:')
print(f' Años válidos (2019-2024): {anios_validos.sum():,} registros')
if anios_invalidos > 0:
    print(f' Años inválidos: {anios_invalidos:,} registros')
    print(f'      Años problemáticos: {sri_dataset[~anios_validos]["ANIO"].unique()}')
else:
    print(f' No se encontraron años fuera de rango')

# 2. Validar valores negativos en ventas
ventas_negativas = (sri_dataset['TOTAL_VENTAS'] < 0).sum()
compras_negativas = (sri_dataset['TOTAL_COMPRAS'] < 0).sum()
print(f'\n 2 VALIDACIÓN DE VALORES NEGATIVOS:')
if ventas_negativas > 0:
    print(f' Ventas negativas: {ventas_negativas:,} registros ({ventas_negativas/len(sri_dataset)*100:.2f}%)')
else:
    print(f' No hay ventas negativas')
if compras_negativas > 0:
    print(f' Compras negativas: {compras_negativas:,} registros ({compras_negativas/len(sri_dataset)*100:.2f}%)')
else:
    print(f' No hay compras negativas')

# 3. Validar meses
meses_validos = sri_dataset['MES'].between(1, 12)
meses_invalidos = (~meses_validos).sum()
print(f'\n 3 VALIDACIÓN DE MESES:')
if meses_invalidos > 0:
    print(f' Meses inválidos (fuera de 1-12): {meses_invalidos:,}')
else:
    print(f' Todos los meses están en rango válido (1-12)')

# 4. Validar provincias vacías
provincias_vacias = sri_dataset['PROVINCIA'].isna().sum()
print(f'\n 4 VALIDACIÓN DE PROVINCIAS:')
if provincias_vacias > 0:
    print(f' Provincias vacías: {provincias_vacias:,} registros')
else:
    print(f' Todas las provincias están completas')
print(f' Provincias únicas: {sri_dataset["PROVINCIA"].nunique()}')

# 5. Estadísticas de duplicados
print(f'\n 5 VALIDACIÓN DE DUPLICADOS:')
print(f' Duplicados ya eliminados en paso anterior')

# 6. Completitud por columna clave
print(f'\n 6 COMPLETITUD DE COLUMNAS CLAVE:')
for col in ['ANIO', 'MES', 'PROVINCIA', 'TOTAL_VENTAS', 'TOTAL_COMPRAS']:
    completitud = sri_dataset[col].notna().mean() * 100
    print(f'{col}: {completitud:.2f}%')
```

```python

# DETECCIÓN DE OUTLIERS (Valores Atípicos)

def detectar_outliers_iqr(df, columna):
    """
    Detecta outliers usando el método IQR (Rango Intercuartílico)
    Outliers = valores fuera de [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
    """
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]
    return outliers, limite_inferior, limite_superior

print('-'*60)
print('  ANÁLISIS DE OUTLIERS (Valores Atípicos)')


# Analizar outliers en TOTAL_VENTAS
sri_dataset['TOTAL_VENTAS'] = pd.to_numeric(sri_dataset['TOTAL_VENTAS'], errors='coerce')
outliers_ventas, lim_inf, lim_sup = detectar_outliers_iqr(sri_dataset, 'TOTAL_VENTAS')

print(f'\n OUTLIERS EN TOTAL_VENTAS:')
print(f'   Límite inferior: ${lim_inf:,.2f}')
print(f'   Límite superior: ${lim_sup:,.2f}')
print(f'   Outliers detectados: {len(outliers_ventas):,} ({len(outliers_ventas)/len(sri_dataset)*100:.2f}%)')

if len(outliers_ventas) > 0:
    print(f'\n  Top 5 outliers (ventas más altas):')
    top_outliers = outliers_ventas.nlargest(5, 'TOTAL_VENTAS')[['PROVINCIA','CODIGO_SECTOR_N1', 'ANIO', 'TOTAL_VENTAS']]
    for idx, row in top_outliers.iterrows():
        print(f' • {row["PROVINCIA"]} - {row["CODIGO_SECTOR_N1"]} ({row["ANIO"]}): ${row["TOTAL_VENTAS"]:,.0f}')

# Visualizar con boxplot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Boxplot de ventas (escala normal)
sri_dataset.boxplot(column='TOTAL_VENTAS', ax=axes[0])
axes[0].set_title('Distribución de TOTAL_VENTAS (con outliers)', fontweight='bold')
axes[0].set_ylabel('USD')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: f'${x/1e6:.0f}M'))

# Boxplot de ventas (escala logarítmica para ver mejor)
sri_dataset_sin_ceros = sri_dataset[sri_dataset['TOTAL_VENTAS'] > 0]
sri_dataset_sin_ceros.boxplot(column='TOTAL_VENTAS', ax=axes[1])
axes[1].set_yscale('log')
axes[1].set_title('Distribución de TOTAL_VENTAS (escala log)', fontweight='bold')
axes[1].set_ylabel('USD (escala logarítmica)')

plt.tight_layout()
plt.show()

print(f'\n INTERPRETACIÓN:')
print(f'   Los outliers son normales en datos fiscales. Representan grandes empresas')
print(f'   o sectores con operaciones de alto volumen (ej: distribuidoras, exportadoras).')
print(f'   NO los eliminamos porque son datos válidos, pero los consideramos en el análisis.')
```

# Enriquecimiento con Region geografica

```python
# Mapa de provincias a regiones del Ecuador
REGIONES = {
    'ESMERALDAS': 'Costa', 'MANABI': 'Costa', 'GUAYAS': 'Costa',
    'SANTA ELENA': 'Costa', 'LOS RIOS': 'Costa', 'EL ORO': 'Costa',
    'SANTO DOMINGO DE LOS TSACHILAS': 'Costa',
    'CARCHI': 'Sierra', 'IMBABURA': 'Sierra', 'PICHINCHA': 'Sierra',
    'COTOPAXI': 'Sierra', 'TUNGURAHUA': 'Sierra', 'BOLIVAR': 'Sierra',
    'CHIMBORAZO': 'Sierra', 'CANAR': 'Sierra', 'AZUAY': 'Sierra', 'LOJA': 'Sierra',
    'SUCUMBIOS': 'Amazonia', 'NAPO': 'Amazonia', 'ORELLANA': 'Amazonia',
    'PASTAZA': 'Amazonia', 'MORONA SANTIAGO': 'Amazonia', 'ZAMORA CHINCHIPE': 'Amazonia',
    'GALAPAGOS': 'Insular'
}

sri_dataset['REGION'] = sri_dataset['PROVINCIA'].map(REGIONES)

# Verificar provincias sin region asignada
sin_region = sri_dataset[sri_dataset['REGION'].isna()]['PROVINCIA'].unique()
if len(sin_region) > 0:
    print(f'Provincias sin region asignada: {sin_region}')
else:
    print('Todas las provincias mapeadas correctamente')

display(sri_dataset.describe(include='all'))
completeness_region = sri_dataset['REGION'].notna().mean() * 100
print(f'Completitud de REGION: {completeness_region:.2f}%')
```

# Enriquecimiento con nombre del Sector Economico CIIU

```python
# Clasificacion CIIU 4.0 Nivel 1 -- fuente: INEC Ecuador
SECTORES_CIIU = {
    'A': 'Agricultura y Ganaderia',
    'B': 'Explotacion de Minas y Canteras',
    'C': 'Manufactura',
    'D': 'Electricidad y Gas',
    'E': 'Agua y Saneamiento',
    'F': 'Construccion',
    'G': 'Comercio al por Mayor y Menor',
    'H': 'Transporte y Almacenamiento',
    'I': 'Alojamiento y Restaurantes',
    'J': 'Informacion y Comunicacion',
    'K': 'Actividades Financieras y de Seguros',
    'L': 'Actividades Inmobiliarias',
    'M': 'Actividades Profesionales y Tecnicas',
    'N': 'Actividades Administrativas',
    'O': 'Administracion Publica',
    'P': 'Educacion',
    'Q': 'Salud y Asistencia Social',
    'R': 'Arte y Entretenimiento',
    'S': 'Otras Actividades de Servicio',
    'T': 'Hogares como Empleadores',
    'U': 'Organizaciones Extraterritoriales'
}

# Extraer la letra inicial del codigo CIIU y mapear al nombre
sri_dataset['SECTOR_LETRA'] = sri_dataset['CODIGO_SECTOR_N1'].str[0]
sri_dataset['SECTOR_NOMBRE'] = sri_dataset['SECTOR_LETRA'].map(SECTORES_CIIU).fillna('Sin Clasificar')

print('Sectores unicos encontrados:', sri_dataset['SECTOR_NOMBRE'].nunique())
display(sri_dataset.describe(include='all'))
completeness_sector = sri_dataset['SECTOR_NOMBRE'].notna().mean() * 100
print(f'Completitud de SECTOR_NOMBRE: {completeness_sector:.2f}%')
```

# Creacion de variables temporales derivadas

```python
# Agregar trimestre y nombre del mes
MESES = {1:'Enero',2:'Febrero',3:'Marzo',4:'Abril',5:'Mayo',6:'Junio',
          7:'Julio',8:'Agosto',9:'Septiembre',10:'Octubre',11:'Noviembre',12:'Diciembre'}

sri_dataset['TRIMESTRE'] = pd.cut(
    sri_dataset['MES'], bins=[0,3,6,9,12], labels=['Q1','Q2','Q3','Q4']
)
sri_dataset['MES_NOMBRE'] = sri_dataset['MES'].map(MESES)

# Calcular MARGEN = diferencia entre total ventas y total compras
sri_dataset['MARGEN'] = sri_dataset['TOTAL_VENTAS'] - sri_dataset['TOTAL_COMPRAS']

print('Variables temporales y MARGEN creados correctamente')
display(sri_dataset.head())
display(sri_dataset.describe(include='all'))
```

# Analisis matematico por provincia y año

```python
# Calcular ventas totales por provincia y año
ventas_prov_año = (
    sri_dataset.groupby(['PROVINCIA','ANIO'])['TOTAL_VENTAS']
    .sum()
    .reset_index()
)
ventas_prov_año['TOTAL_VENTAS_M'] = (ventas_prov_año['TOTAL_VENTAS'] / 1e6).round(2)
display(ventas_prov_año.head(20))
```

```python
# Calcular crecimiento porcentual de ventas 2019 -> 2024 por provincia
pivot_prov = ventas_prov_año.pivot(index='PROVINCIA', columns='ANIO', values='TOTAL_VENTAS')
pivot_prov['CRECIMIENTO_%'] = ((pivot_prov[2024] - pivot_prov[2019]) / pivot_prov[2019] * 100).round(1)
pivot_prov['VOLUMEN_2024_M'] = (pivot_prov[2024] / 1e6).round(1)

print('Crecimiento por provincia 2019->2024:')
display(pivot_prov[['VOLUMEN_2024_M','CRECIMIENTO_%']].sort_values('CRECIMIENTO_%', ascending=False))
```

```python
# Calcular crecimiento por sector economico
ventas_sector_año = (
    sri_dataset.groupby(['SECTOR_NOMBRE','ANIO'])['TOTAL_VENTAS']
    .sum()
    .reset_index()
)
pivot_sector = ventas_sector_año.pivot(index='SECTOR_NOMBRE', columns='ANIO', values='TOTAL_VENTAS')
pivot_sector['CRECIMIENTO_%'] = ((pivot_sector[2024] - pivot_sector[2019]) / pivot_sector[2019] * 100).round(1)
pivot_sector['VOLUMEN_2024_M'] = (pivot_sector[2024] / 1e6).round(1)

print('Crecimiento por sector 2019->2024:')
display(pivot_sector[['VOLUMEN_2024_M','CRECIMIENTO_%']].sort_values('CRECIMIENTO_%', ascending=False))
```

```python
# Funcion para estimar ventas de una provincia dado un año

def estimar_ventas(provincia, anio):
    provincia = provincia.upper()
    filtro = sri_dataset[(sri_dataset['PROVINCIA'] == provincia) & (sri_dataset['ANIO'] == anio)]
    if filtro.empty:
        return f'No hay datos para {provincia} en {anio}'
    return {
        'provincia': provincia,
        'anio': anio,
        'total_ventas': filtro['TOTAL_VENTAS'].sum(),
        'total_compras': filtro['TOTAL_COMPRAS'].sum(),
        'margen': filtro['MARGEN'].sum(),
        'exportaciones': filtro['EXPORTACIONES'].sum(),
        'num_registros': len(filtro)
    }

# Ejemplo de uso
print(estimar_ventas('PICHINCHA', 2024))
print(estimar_ventas('GUAYAS', 2024))
```

```python
# Grafico 8: Mapa de oportunidades - sectores por volumen vs crecimiento
# Sectores con ALTO crecimiento y BAJO volumen = nichos de mercado
crec_sector = pivot_sector[['VOLUMEN_2024_M','CRECIMIENTO_%']].dropna()

p35_vol  = crec_sector['VOLUMEN_2024_M'].quantile(0.35)
p65_crec = crec_sector['CRECIMIENTO_%'].quantile(0.65)

nichos = crec_sector[
    (crec_sector['VOLUMEN_2024_M'] <= p35_vol) &
    (crec_sector['CRECIMIENTO_%'] >= p65_crec)
]
print('Sectores con mayor oportunidad (nicho):')
display(nichos)

fig, ax = plt.subplots(figsize=(14, 9))
ax.scatter(crec_sector['VOLUMEN_2024_M'], crec_sector['CRECIMIENTO_%'],
           s=100, alpha=0.6, color='steelblue', zorder=3)
ax.scatter(nichos['VOLUMEN_2024_M'], nichos['CRECIMIENTO_%'],
           s=180, color='gold', edgecolor='orange', zorder=4, label='Nicho / Oportunidad')
for sector, row in crec_sector.iterrows():
    ax.annotate(sector, (row['VOLUMEN_2024_M'], row['CRECIMIENTO_%']),
                fontsize=8, xytext=(4,4), textcoords='offset points')
ax.axvline(p35_vol,  color='gray', linestyle='--', alpha=0.5)
ax.axhline(p65_crec, color='gray', linestyle=':',  alpha=0.5)
ax.set_title('Mapa de Oportunidades por Sector Economico', fontsize=14, fontweight='bold')
ax.set_xlabel('Ventas 2024 (Millones USD)')
ax.set_ylabel('Crecimiento 2019-2024 (%)')
ax.legend()
plt.tight_layout()
plt.show()
```

**¿Qué muestra este gráfico?**
- Scatter plot donde cada punto es un sector económico
- Eje X: Volumen de ventas 2024 (tamaño del mercado)
- Eje Y: Crecimiento 2019-2024 (dinamismo)
- Puntos dorados: Sectores identificados como "nichos" u oportunidades

**Cómo leerlo:**
- **Arriba-Izquierda (NICHOS)**: Alto crecimiento + Bajo volumen = Mercados emergentes con menos competencia
- **Arriba-Derecha**: Alto crecimiento + Alto volumen = Mercados grandes y en expansión (muy competidos)
- **Abajo-Izquierda**: Bajo crecimiento + Bajo volumen = Mercados pequeños y estancados
- **Abajo-Derecha**: Bajo crecimiento + Alto volumen = Mercados maduros y saturados

**Hallazgos clave:**
1. **Sectores NICHO identificados** (ejemplos típicos):
   - Información y Comunicación (J): Tecnología, software, telecomunicaciones
   - Actividades Profesionales (M): Consultoría, servicios técnicos
   - Salud privada (Q): Clínicas, laboratorios
   
   **Por qué son oportunidades:**
   - Crecen >50% pero aún son pequeños
   - Menos jugadores establecidos = menor competencia
   - Tendencia al alza sostenida

2. **Sectores grandes pero lentos** (abajo-derecha):
   - Comercio tradicional: Grande pero crece poco
   - Manufactura básica: Estable pero saturada

3. **El tamaño importa según tu objetivo**:
   - Startup o PYME: busca nichos (arriba-izquierda)
   - Gran empresa: puede competir en sectores grandes (derecha)

**Implicaciones estratégicas:**
- **Para emprendedores**: Enfócate en los puntos dorados. Son mercados donde puedes entrar con menos capital y enfrentar menos gigantes.
- **Para inversionistas**: Los nichos ofecen mayor retorno potencial (están creciendo rápido).
- **Tendencia**: Ecuador se está moviendo hacia economía de servicios (los nichos suelen ser servicios, no productos).

**Ejemplo práctico:**
Si tienes $50,000 para invertir:
-  Mal: Abrir una tienda de ropa (Comercio G) → Mercado saturado
-  Bueno: Ofrecer servicios de marketing digital (Información J) → Nicho en crecimiento
